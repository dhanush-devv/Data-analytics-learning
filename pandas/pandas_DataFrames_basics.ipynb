```python
import pandas as pd
```

```python
#Data Frame
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
df
```

```python
# Columns
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"],index=["x","y","z"])
df
```

```python
# Head Gives top element
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"])
df.head()
```

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"])
df.head(1)
```

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"])
df.head(2)
```
