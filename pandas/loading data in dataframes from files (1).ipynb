```python
bios
```

```python
def categorize_athelete(row):
    if row['height_cm']<175 and row['weight_kg']<70 :
        return 'LightWeight'
    elif row['height_cm']<185 and row['weight_kg']<=80 :
        return 'MiddleWeight'
    else:
        return 'HeavyWeight'
bios['Category']=bios.apply(categorize_athelete,axis=1)
```

# Merging and Concatanation

```python
noc=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv')
```

```python
noc.head()
```

```python
bios.head()
```

```python
bios_new=pd.merge(bios,noc,left_on='born_country',right_on='NOC',how='left')
```

```python
bios_new
```

```python
bios_new.rename(columns={'region':'born_country'},inplace=True)
```

```python
bios_new
```

```python
bios_new[bios_new['NOC_x']!=bios_new['born_country']]
```
